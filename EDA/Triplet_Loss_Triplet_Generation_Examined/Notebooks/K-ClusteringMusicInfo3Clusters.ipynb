```python
# The csv file is imported as a data frame through Pandas
import pandas as pd
```

```python
Musicinfo_df = pd.read_csv("music_info_chosenattributes.csv")
Musicinfo_df.head(5)
```

```python
# Get rid of two unnecessary columns
Reduced_Musicinfo_df = Musicinfo_df.drop(columns = ['similarity_to_MrBrightside','X_vector'])
```

```python
Reduced_Musicinfo_df.head(5)
```

```python
# Import the K-means class from the clusters module in Scikit-learn
from sklearn.cluster import KMeans
```

```python
# Define the inputs we will use for our K-means clustering algorithm
X = Reduced_Musicinfo_df[['danceability','energy','key','mode','valence','tempo']].copy()
```

```python
# Finally, let's fit/assign clusters to each of the data points in the data frame 
kmeans = KMeans(n_clusters=3)
y = kmeans.fit_predict(Reduced_Musicinfo_df[['danceability','energy','key','mode','valence','tempo']])
Reduced_Musicinfo_df['Cluster'] = y
print(Reduced_Musicinfo_df.head())
```

```python
Reduced_Musicinfo_df.head(4)
```

```python
# write the previous data frame to a csv file 
Reduced_Musicinfo_df.to_csv("3clusters_musicinfo.csv",index=False)
```
